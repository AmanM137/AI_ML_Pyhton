## tips on basis of total bill in tips datasets usin SLR from scratch

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
```

```python
tips.head()
```

```python
y=tips['tip']
X1=tips['total_bill']
```

```python
y
```

```python
plt.hist(y,ec='k',bins=30)
plt.grid(alpha=0.5)
plt.title("Tip Distribution")
plt.xlabel("Tip")
plt.ylabel("Frequency")
plt.xticks(np.arange(0,11,1))
plt.show()
```

```python
X1
```

```python
plt.hist(X1,ec='k',bins=30)
plt.grid(alpha=0.5)
plt.title("Total Bill Distribution")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.xticks(np.arange(0,55,5))
plt.show()
```

```python
r=np.corrcoef(X1,y)[0,1]
r
```

```python
plt.plot(X1,y,"bo",mfc="white",alpha=0.5,label=f"r:{r:.2f}")
plt.xlabel("Total Bill")
plt.ylabel("Tips")
plt.title("Total Bill vs Tips")
plt.grid(alpha=0.5,ls='--')
plt.legend()
plt.show()
```

```python
theta_1 =  ( (y.mean() - y)*(X1.mean() - X1) ).sum() / ( ( X1.mean() - X1)**2 ).sum()
theta_0 =  y.mean() - theta_1 * X1.mean()
print("Coef_ / Slope = ", theta_1)
print("Intercept = ", theta_0)
```

```python
y_hat = theta_0 + theta_1 * X1
```

```python
plt.plot(X1, y, "bo", mfc="w", alpha=0.5, label=f"r = {r:0.2f}")
plt.plot(X1,y_hat, "r-", label="prediction line")
plt.xlabel("Total Bill")
plt.ylabel("Tips")
plt.title("Total Bill vs Tips")
plt.grid(alpha=0.5, ls='--')
plt.legend()
plt.show()
```
